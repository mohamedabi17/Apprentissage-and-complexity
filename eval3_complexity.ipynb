```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from scipy.stats import pearsonr 
```

```python

X_train = np.array([8, 8.9, 10.7, 12.8, 15, 18]).reshape(-1, 1)
Y_train = np.array([55, 120, 275, 610, 1365, 3050])
```

```python


# Given data
X_train = np.array([8, 8.9, 10.7, 12.8, 15, 18]).reshape(-1, 1)
Y_train = np.array([55, 120, 275, 610, 1365, 3050])


def linear_model(X):
    return X

def log_model(X):
    return np.log(X)

def exp_model(X):
    return np.exp(X)

def power_model(X, power):
    return np.power(X, power)


linear_preds = linear_model(X_train)
log_preds = log_model(X_train)
exp_preds = exp_model(X_train)


power_preds = power_model(X_train, power=2)

corr_linear = pearsonr(linear_preds.flatten(), Y_train)[0]
corr_log = pearsonr(log_preds.flatten(), Y_train)[0]
corr_exp = pearsonr(exp_preds.flatten(), Y_train)[0]
corr_power = pearsonr(power_preds.flatten(), Y_train)[0]


print(f"Correlation for Linear Model: {corr_linear}")
print(f"Correlation for Logarithmic Model: {corr_log}")
print(f"Correlation for Exponential Model: {corr_exp}")
print(f"Correlation for Power Model (puissance): {corr_power}")
```

```python

# Visualisation des données
plt.scatter(X_train, Y_train, color='blue', label='Données d\'apprentissage')
plt.title('Relation entre X et Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
degree = 2 
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X_train)
model = LinearRegression()
model.fit(X_poly, Y_train)
```

```python
X_test = np.array([20]).reshape(-1, 1)
X_test_poly = poly_features.transform(X_test)
Y_pred = model.predict(X_test_poly)
```

```python

print(f"Prédiction pour X = 20 : {Y_pred[0]}")
```

```python

Y_observed = 4449.926785048952  
error = Y_observed - Y_pred[0]
print(f"Erreur de prédiction : {error}")

```

```python

X_range = np.linspace(min(X_train), max(X_train), 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
Y_range_pred = model.predict(X_range_poly)

```

```python
plt.scatter(X_train, Y_train, color='blue', label='Données d\'apprentissage')
plt.plot(X_range, Y_range_pred, color='red', label='Modèle d\'ajustement')
plt.title('Régression polynomiale entre X et Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
